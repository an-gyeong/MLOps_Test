#### 2) Dockerfile 문법
- 아래 예시로 작성된 Dockerfile 내 지시어의 의미는 다음과 같습니다.
    - FROM : 어떤 베이스 이미지를 사용할건지
    - LABEL : 이미지의 메타데이터라고 생각하면 되고, 필수로 작성할 필요 없음
    - expose: 문서용도로 사용 실제 동작하지는 않음
        - expose를 한다고해서 해당 port가 publshing 되지는 않습니다. 그저 dockerfile에서 명시적으로 이런 포트를 사용한다고 알려주는 용도
    - CMD : 해당 이미지를 통해 컨테이너를 실행할 때 어떤 명령어를 수행할지 결정

FROM node: 12-alpine  
LABEL  
RUN apk add --no-cache python3 g++ make  
WORKDIR /app  
COPY . .  
RUN yarn install --production  
EXPOSE 80  
CMD ["node", "src/index.js"]

#### 1) Dockerfile 문법 
1) Dockerfile 만들기
    - 로컬 pc경로에 Dockerfile 이라는 이름의 빈 파일을 하나 생성합니다
    
2) 문법
- FROM

    #예시  
    FROM ubuntu  
    FROM ubuntu:20.04  
    FROM nginx:latest AS ngx  
    

- LABEL
    - 이미지의 메타 데이터라고 보면 되고 옵션이라 반드시 사용해야 하는건 아님
    
    #예시  
    LABEL maintainer="xxx@gmail.com"  
    LABEL version="1.0.0"  
    LABEL description="Write description"  
    

- COPY
    - 복사하고 싶은 경로< src>의 파일 혹은 디렉토리를 최종 경로< dest>에 복사하는 명령어

    COPY < src>...< dest>  

    #예시  
    COPY a.txt /some-directory.txt  
    COPY my-directory /some-directory-2  
    

- RUN
    - 명시한 커맨드를 도커 컨테이너에서 실행하는 것을 명시하는 명령어
    
    RUN < command>  
    RUN ["executable-command", "parameter1", "parameter2"]  

    #예시  
    RUN pip install torch  
    RUN pip install -r requirments.txt  
    
    RUN apt-get update  
    RUN apt-get upgrade -y  
    RUN apt-get clean && apt-get autoremocve -y  
    

- CMD
    - 명시한 커맨드를 통해 컨테이너가 시작될 때, 실행하는 것을 명시하는 명령어
    - 하나의 docker image에서는 하나의 CMD만 실행할 수 있다는 점에서 RUN명령어와 다름

    CMD < command>  
    CMD ["executable-command", "parameter1", "parameter2"]  
    CMD ["parameter1", "parameter2"] # ENTRYPOINT와 함께 사용될때  

    #예시  
    CMD python main.py  
    

- WORKDIR
    - 작성된 명령어들을 컨테이너 내 어떤 디렉토리에서 수행할 것인지를 명시하는 명령어

    WORKDIR /path/to/workdir  

    #예시  
    WORKDIR /home/demo  
    

- EXPOSE
    - 컨테이너에서 사용할 port/protocol을 지정할 수 있음
    - 단, EXPOSE로 지정한다고해서 해당 port가 publiching되지는 않음. 그저 dockerfile에서 명시적으로 이런 포트를 사용한다고 알려주는 용도

    EXPOSE < port>  
    EXPOSE < port>/< protocol>  

    #예시  
    EXPOSE 8080  
    

- ENV
    - 컨테이너 내부에서 지속적으로 사용할 environment variable값을 설정하는 명령어

    ENV < key>< value>  
    ENV < key>=< value>  

    #예시  
    #default 언어 설정  
    ENV locale-gen ko_KR.UTF-8  
    ENV LANG ko_KR.UTF-8  
    ENV LANGUAGE ko_KR.UTF-8  
    ENV LC_ALL ko_KR.UTF-8

##### 1) Docker Compose
- 공식 홈페이지 문서 : https://docs.docker.com/compose/features-uses/
- 기존에 명령어를 통해서 컨테이너를 실행하는 일련의 행위들을 도커 컴포즈에서 제공하는 docker-compose.yml 이라는 yaml 파일을 이용해서 명시적으로 관리가 가능합니다.
- 단일 서버에서 여러 컨테이너를 프로젝트 단위로 묶어서 관리가 가능합니다.
- Docker compose 사용 이점
    - 프로젝트 단위로 도커 네트워크와 볼륨 관리
    - 프로젝트 내 서비스 간 의존성 정의 가능
        - 서비스 의존성이란? 어떤 서비스가 실행되기전에 먼저 실행되어야 할 서비스등을 정의할 수 있음
        - 손쉬운 컨테이너 수평 확장
- 2023년 6월말부터 Compose V1은 더이상 지원되지 않으며 Docker Desktop버전에서도 삭제된다고 합니다.
    - Docker Compose CLI 플러그인을 사용하거나, Docker Desktop에서 Docker Compose V2 사용 설정을 활성화하여 ComposeV2로 전환해야 합니다.



##### 2) Project / Services / Containers
1. Project(프로젝트)
- Docker compose에서 다루는 워크스페이스 단위입니다.
- 함께 관리하는 서비스 컨테이너의 묶음이라고 할 수 있습니다.
- 프로젝트 단위로 기본 Docker network가 생성됩니다.
- docker-compose.yml 하나가 Project를 명세한다고 볼 수 있습니다.

2. Service(서비스)
- Docker compose에서 컨테이너를 관리하기 위한 최소 단위입니다.
- 예를 들어 서비스에서 scale을 통해 서비스 컨테이너의 수 등을 확장 가능 할수 있습니다.

3. Container(컨테이너)
- Service를 통해 기본적으로 관리되는 것을 컨테이너라고 합니다.

##### 3) docker-compose.yml
- version, services, networks, volumes 총 4개의 최상위 옵션을 제공합니다.  
  

1) version
- 2021년 11월 기준 버전 3.9가 최신입니다.
- 가능한 최신 버전 사용을 권장, docker engine과 docker compose 버전에 따른 호환성 매트릭스가 있으니 참조해서 사용
    - 호환성 매트릭스 : https://docs.docker.com/compose/compose-file/compose-file-v3/
- 버전 3부터 Docker Swarm과 호환이 가능
    - Swarm 서비스를 docker-compose.yml로 정의해서 사용 가능
- Docker Swarm
    - 여러 서비스를 기반으로 Swarm cluster를 형성하여 컨테이너를 관리하는 container ochestration 시스템
    - Kuvernetes와 동일 목적으로 만들어졌지만 현재 많은 인기를 끌지는 못함
    - 주의점 : Docker Swarm과 관련된 기능들이 추가되면서 Docker compose 에서 사용되는 옵션인지, 아니면 Docker Swarm 에서 사용되는 옵션인지 구분해야함
